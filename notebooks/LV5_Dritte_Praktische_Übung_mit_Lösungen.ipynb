Verwenden Sie die Funktion aus der LV5 um die H√§ufigkeiten der W√∂rter in dieser Datei zu z√§hlen und auszugeben.

```python
# F√ºgen Sie hier Ihren Code hier ein
def word_frequency(text):
  words = text.split()
  count_dict = {}
  for word in words:
    if word in count_dict.keys():
      count_dict[word] += 1
    else:
      count_dict[word] = 1
  return count_dict

word_frequency(text2)
```

Schreiben Sie nun eine Funktion, welche die Anzahl der Eigennamen im Text z√§hlt und ausgibt. Verwenden Sie dazu spaCy. Dazu muss erst das richtige Modell f√ºr spaCy geladen werden.

Danach muss wiederum die Bibliothek importiert werden.

```python
import spacy
nlp = spacy.load("de_core_news_sm")

# F√ºgen Sie hier Ihren Code hier ein

def count_named_entities(text):
  doc = nlp(text)
  nes = {}
  for ent in doc.ents:
    if ent.text in nes.keys():
      nes[ent.text] += 1
    else:
      nes[ent.text] = 1
  # Set damit Doppelnennungen nicht zwei Mal gez√§hlt werden
  return nes

result = count_named_entities(text2)
print("Die Anzahl der im Text gefundenen Eigenname ist", len(result.keys()))
print("Die gefundenen Eigennamen und ihre H√§ufigkeiten sind")
print(result)
```
