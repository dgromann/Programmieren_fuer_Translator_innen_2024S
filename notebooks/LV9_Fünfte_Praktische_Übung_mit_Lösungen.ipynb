Schreiben Sie dazu eine Funktion, welche √ºberpr√ºft ob f√ºr ein Token ein Synset in WordNet gefunden wird oder nicht (`wordnet.synsets(token)`). Z√§hlen Sie dabei wie viele Tokens ein Synset aufweisen und berechnen Sie die Prozent dieser Anzahl im Vergleich zur Gesamtanzahl der Tokens.

Bitte beachten Sie, dass der vorangehende Unterstrich _ bei den Tokens entfernt werden sollte, nicht jedoch jedes Token einen Unterstrich hat.  Achten Sie bitte auch darauf, dass der Unterstrich eine ganz eigene Encodierung hat und direkt aus der Ausgabe des Vokabulars kopiert werden sollte (nicht als Unterstrich auf der Tastatur eingegeben werden sollte).

Wie unterscheidet sich die Prozentzahl der gefundenen Tokens mit oder ohne Entfernung des Unterstrichs?

```python
# F√ºgen Sie hier Ihren Code hier ein
def find_english_tokens_with_underscore(vocab):
  tokens_in_wordnet = []
  for token in vocab.keys():
    if wordnet.synsets(token):
      tokens_in_wordnet.append(token)
  return tokens_in_wordnet

def find_english_tokens_without_underscore(vocab):
  tokens_in_wordnet = []
  for token in vocab.keys():
    token = token.replace("‚ñÅ", "")
    if wordnet.synsets(token):
      tokens_in_wordnet.append(token)
  return tokens_in_wordnet

english_with_underscore = find_english_tokens_with_underscore(vocabulary)
english_without_underscore = find_english_tokens_without_underscore(vocabulary)

print(f"Die Gesamtanzahl der Tokens in WordNet ohne Entfernung des Unterstrichs betr√§gt {round((len(english_with_underscore)/len(vocabulary))*100,2)}% aller Tokens.")
print(f"Diese sind {english_with_underscore} \n")
print(f"Die Gesamtanzahl der Tokens in WordNet mit Entfernung des Unterstrichs betr√§gt {round((len(english_without_underscore)/len(vocabulary))*100,2)}% aller Tokens.")
print(f"Diese sind {english_without_underscore}")
```
